# Image Scaling
This notebook contains examples of downsampling xView images.

```python
import data_utilities.aug_util as aug
import data_utilities.wv_util as wv
import matplotlib.pyplot as plt
import numpy as np
import csv
import json as json
```

```python
data_path = "dataset/"
image_path = "train_images/"
img_num = "104.tif"

path = data_path+image_path+img_num
arr = wv.get_image(path)

plt.figure(figsize=(20,20))
plt.axis('off')
plt.imshow(arr)
plt.show()
```

```python
downsampled_img = wv.resize(arr,5)

plt.figure(figsize=(20,20))
plt.axis('off')
plt.imshow(downsampled_img)
plt.show()
```

# Original Resolution
The images below shows the original resolution of the xView Dataset

```python
#We can chip the image into 500x500 chips
tile_shape = (500,500)
c_img = wv.chip_image_only(img = arr, shape=(500,500))
```

# Downsized Resolution
The images below shoes the new downsized resolution, and new chip calculations

```python
orig_res = 0.3
new_resolution=5
orig_tile_shape = (500,500)
scale_percent = orig_res/new_resolution
chip_width = int(orig_tile_shape[0] * scale_percent)
chip_height = int(orig_tile_shape[1] * scale_percent)
tile_shape = (chip_width, chip_height)
```

```python
# Extract chipped images
d_c_img = wv.chip_image_only(img = downsampled_img,shape=tile_shape)
```

```python
for ind in range(2):
    plt.figure(figsize=(5,5))
    plt.imshow(d_c_img[ind])
    plt.axis('off')
    plt.show()
```
